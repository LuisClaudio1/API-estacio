```python
import pandas as pd
# Biblioteca basedosdados está apresentando erro para dowmload
# import basedosdados as bd
```

```python
# Carregando os dados via arquivo local
df = pd.read_csv("./dados/pesquisa_agrícola_municipal_2010.csv")

# Breve visualização dos dados
df.head()
```

```python
# Criando uma busca dentro da tabela

def pesquisa(produto, uf, baixar=2):
    
    # Criando a busca apartir do nome do produto
    df_busca1 = df[df['produto'].str.contains(produto, case=False, regex=False)]
    
    # Buscado o UF apartir da busca anterior
    df_busca_final = df_busca1[df_busca1['sigla_uf'].str.contains(uf, case=False, regex=False)]

    # Verificando o retorno da busca realizada
    if len(df_busca_final) == 0 or len(df_busca_final) == None: # Caso não tenha valor para a busca realizada
        print("Busca não encontrada!")
    
    else:
        if baixar == '1' or baixar == 1:    # Baixando o arquivo em formato .xlsx
            df_busca_final.to_excel(f"pesquisas/pesquisa_{produto}_{uf}.xlsx")
            print("Arquivo foi salvo na pasta 'pesquisas'")
        
        elif baixar =='2' or baixar ==2:    # Apenas imprimindo a busca na tela
            return df_busca_final
        
        else:                               # Caso a opção informada sejá fora do esperado
            print("Tente novamente, informando um numero que seja 1 ou 2.\n Sendo 1 para baixar o arquivo e 2 para apenas aparecer na tela!") 
        
```

```python
# Realizando a busca do usuario
nome_produto = input("Qual o nome do produto: ")
sigla_uf = input("Qual a sigla do estado: ")
baixar = input("Deseja baixar a busca? (1 realiza o download e 2 apenas retora a busca): ")
```

```python
pesquisa(nome_produto, sigla_uf, baixar)
```
